```python
import numpy as np
import torch
import torch 
import torch.utils.data
from torch.utils.data import Dataset, DataLoader
import torchvision
import matplotlib.gridspec as gridspec
```

# Formalise the loss_500.npy

```python
loss_ite = np.load('/home/research/yuqijing/sem2project/Diffusion/LossLog/LossI_500.npy')
# 500 epochs, 128 batch size, 50_000 images
# 391 iterations, 
print(loss_ite.shape)

loss_epo = np.array([])
N = 391
for i in range(502):
    item = loss_ite[i*N]
    loss_epo = np.append(loss_epo, item)
    
loss_epo 
loss = loss_epo[:501]
print(loss[0])
print(len(loss))
np.save('LossE_500.npy', loss)



 
import matplotlib.pyplot as plt    
import matplotlib.gridspec as gridspec

fig, ax = plt.subplots(figsize=(16, 8), facecolor='white')
# plt.figure(figsize=(16, 8))
plt.plot(loss, color='darkred', linestyle='-.', linewidth=2)
leg = plt.legend(["%s" % p for p in ['CIFAR10']], fontsize=15)
leg.get_frame().set_edgecolor('black')
leg.get_frame().set_linewidth(1)
leg.get_frame().set_boxstyle('square')
plt.xlabel('Epochs', fontsize=15)
plt.ylabel('Training Loss', fontsize=15)
plt.xticks(fontsize=14)
plt.yscale('log')
x = [2, 0.2, 0.02, 0.002]
y = ['2', '0.2', '0.02', '0.002']
plt.yticks(x, y, fontsize=14)
plt.minorticks_off()
ax.tick_params(axis='x', direction='in', length=8)
ax.tick_params(axis='y', direction='in', length=8)
plt.grid(linestyle='-.', axis='y', linewidth=0.3, color='black')
plt.xlim(-2, 510)
plt.savefig('Loss_C_500.png')
plt.show()
    
    

```
